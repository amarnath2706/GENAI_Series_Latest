```python
#Tokenizer working mechanism

from transformers import AutoTokenizer

#load a pretrained tokenizer
#Transformer based bert model
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

#Example Text
text = "I was not happy with the movie"

#Tokenize the text
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

```

```python
#convert tokens to input ID's or Vectors
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input ID's:",input_ids)
```

```python
#Best method to convert the text to vectors
#Encode the text (tokenization + converting to input IDs )
encoded_input = tokenizer(text)
print("Encoded Input:",encoded_input)

#101 - Start of the sentence
#102 - End of the sentence
#attention_mask - for which word we need to give more attention
```

```python
 #Decode the text
decoded_output = tokenizer.decode(input_ids)
print("Decoded_Output:",decoded_output)
```

```python
#Tokenizer working mechanism

from transformers import AutoTokenizer

#load a pretrained tokenizer
#Transformer based bert model
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

#Example Text
text = "I was not happy with the movie"

#Tokenize the text
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

#convert tokens to input ID's or Vectors
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input ID's:",input_ids)

#Best method to convert the text to vectors
#Encode the text (tokenization + converting to input IDs )
encoded_input = tokenizer(text)
print("Encoded Input:",encoded_input)

#Decode the text
decoded_output = tokenizer.decode(input_ids)
print("Decoded_Output:",decoded_output)


```
